```python
import pandas as pd
left = pd.DataFrame({
         'id':[1,2,3,4,5],
         'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
         'subject_id':['sub1','sub2','sub4','sub6','sub5']})

right = pd.DataFrame({    
    'id':[1,2,13,14,5],
         'Name': ['Billy', 'Brian', 'bran', 'bryce', 'Berry'],
         'subject_id':['sub2','sub4','sub3','sub6','sub5']})
print(f'{left}\n{"*"*25}\n{right}')
```

```python
# default inner joining
print(pd.merge(left,right, on='id'))
```

```python
# left joining
print(pd.merge(left,right, on='id', how='left'))
```

```python
# Right joining
print(pd.merge(left,right, on='id', how='right'))
```

```python
# Outer joining or full joining
print(pd.merge(left,right, on='id', how='outer'))
```
